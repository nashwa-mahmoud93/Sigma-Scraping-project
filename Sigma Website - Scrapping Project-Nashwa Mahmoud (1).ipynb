```python
import requests
res = requests.get("https://www.sigma-computer.com/search?search=laptop&submit_search=&route=product%2Fsearch")
res
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, "html.parser")
type(soup)
```

```python
products_divs = soup.find_all("div", attrs={"class": "product-item-container"})
products_divs
```

```python
#item name
name_tag = products_divs[8].find("div", attrs={"class": "caption hide-cont"})
product_name = name_tag.text.replace("\n", " ").strip().replace('"', '""')
product_name
```

```python
#item price after discount
discount_price_tag = products_divs[0].find("span", attrs={"class": "price-new"})
price_after_discount = discount_price_tag.text
price_after_discount
```

```python
#item price before discount
original_price_tag = products_divs[0].find("span", attrs={"class": "price-old"})
price_before_discount = original_price_tag.text
price_before_discount
```

```python
#item status
status_tag = products_divs[0].find("span", class_="stock")
status = status_tag.get_text(strip=True) if status_tag else "Out Of Stock" 
status
```

```python
#item image link 
image_tag = products_divs[0].find('img')
item_image_link = image_tag['src']
item_image_link
```

```python
#item specifications
specifications_tag = products_divs[0].find("div", attrs={"class": "description item-desc"})
item_specifications = specifications_tag.text.replace("\n", " ").strip().replace("\r", " ").replace("\t", " ").replace("\\", " ").replace("  ", " ").replace("  ", " ").replace("  ", " ").replace(",", "")
item_specifications
```

```python
#item link
link_tag = products_divs[0].find("h4").find("a")
if link_tag and "href" in link_tag.attrs:
    product_link = ("https://www.sigma-computer.com/") + link_tag.attrs["href"]
    print(product_link)
else:
    product_link = None
    print("Product link not found")
```

```python
with open("sigma scraping project.csv", mode="w", encoding='utf-8') as fd:
    fd.write("product_name, price_before_discount, price_after_discount, status, image_link, product_link, specifications\n")
    
    for i in range(len(products_divs)):
        name_tag = products_divs[i].find("div", attrs={"class": "caption hide-cont"})
        product_name = name_tag.text.replace("\n", " ").replace('"', '""').strip()
        
        original_price_tag = products_divs[i].find("span", attrs={"class": "price-old"})
        price_before_discount = original_price_tag.text
        
        discount_price_tag = products_divs[i].find("span", attrs={"class": "price-new"})
        price_after_discount = discount_price_tag.text
        
        status_tag = products_divs[i].find("span", class_="stock")
        status = status_tag.get_text(strip=True) if status_tag else "Out Of Stock"
        
        image_tag = products_divs[i].find('img')
        item_image_link = image_tag['src']
        
        
        specifications_tag = products_divs[i].find("div", attrs={"class": "description item-desc"})
        item_specifications = specifications_tag.text.replace("\n", " ").strip().replace("\r", " ").replace("\t", " ").replace("\\", " ").replace("  ", " ").replace("  ", " ").replace("  ", " ").replace(",", "")
        item_specifications
        
        link_tag = products_divs[i].find("h4").find("a")
        if link_tag and "href" in link_tag.attrs:
            product_link = ("https://www.sigma-computer.com/") + link_tag.attrs["href"]
        else:
            "Not Found"
    
        fd.write(f"\"{product_name}\",\"{price_before_discount}\",\"{price_after_discount}\",\"{status}\",\"{item_image_link}\",\"{product_link}\",\"{item_specifications}\"\n")
```
